## Research Question

Our project asks: **Is there a statistically significant relationship between NBA draft position and player career success for players drafted from 2000 to 2018?** We will measure career success using three outcomes: cumulative career **Win Shares**, career average **Player Efficiency Rating (PER)**, and whether a player made at least one **NBA All-Star team**. The main predictor is draft position (pick number), and we will test whether higher picks are associated with stronger outcomes across these success metrics. This is a statistical inference question because we are evaluating the strength and significance of associations rather than building a pure prediction model.


## Background and Prior Work

The NBA Draft is one of the most important ways in which franchises add incoming talent, and draft pick order reflects each organization's perspective on a selected player's projected ability. Higher drafted players have greater expectations of development and thus greater potential for future success because of the extensive scouting resources and tactical planning that franchises invest in selecting the best possible prospect. Due to this, draft pick order is often used as an indicator of a player's perceived worth, so it is valuable to determine whether draft pick order can predict impactful success.

Basketball analytics provide a variety of statistics and measures that allow us to quantify player success based on contributions. Win Shares provide a way to estimate the number of games a franchise has won due to a specific player's effort. Per-minute productivity is summarized in PER (Player Efficiency Rating). Finally, All-Star selections represent a form of elite recognition of players, though they can also be influenced by outside factors including popularity and team success. Using multiple forms of measurement provides a more complete picture of how well a player performs during their career than using one metric alone.

Prior research about draft systems and future success has shown mixed findings across variables. According to a systematic review of North American professional sports, athletes selected in the first round typically perform better in both career length and statistical outcomes.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) In contrast, athletes selected in later rounds are less consistent statistically and therefore harder to evaluate. The study also reports inconsistencies across drafts, which suggests draft position is not a perfect indicator of long-term success.

A similar finding appears in research analyzing the relative value of NBA draft picks. The author concluded that although players selected earlier in the draft perform better on average than those selected later, the relationship is non-linear, meaning being a top pick does not guarantee elite professional performance.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) There is substantial variation in how strongly an individual athlete's performance correlates with draft position.

In addition to draft position and later performance, researchers have studied how general managers make draft decisions. Results indicate that while general managers may value attributes like height or speed more highly when drafting players, these perceptions do not always align with a player's realized contribution over time, such as long-term value or contract outcomes.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) These findings suggest inefficiencies in talent evaluation and show that factors such as player development opportunities, team environment, and access to playing time also shape career outcomes.

Our research builds on existing studies by examining players drafted from 2000 to 2018 and relating draft position to career success measured by Win Shares, Player Efficiency Rating (PER), and All-Star selections. While previous studies have examined draft outcomes, many focus on one success measure or narrower datasets. By combining multiple outcome metrics over a broader modern draft window, we aim to provide a clearer estimate of the association between draft position and future NBA success.

## Hypothesis


We hypothesize that earlier draft picks (lower pick numbers) will, on average, have higher career Win Shares, higher career average PER, and a greater probability of making at least one All-Star team than later picks. We also expect this relationship to be statistically significant but moderate in strength, since player development, injuries, team fit, and opportunity can substantially affect long-term outcomes beyond draft position alone. This expectation is consistent with how teams invest more scouting resources and organizational confidence in higher picks, while still acknowledging substantial variability in NBA career trajectories.

## Data

**Ideal dataset**

Our ideal dataset would include every player selected in the NBA Draft from 2000 through 2018 (about 1,100-1,200 observations across 19 drafts). Each row would represent one drafted player, where draft pick number is the independent variable; career success metrics — total Win Shares, career average PER, seasons played, and All-Star selections — are the dependent variables; and age at draft and minutes played (as a proxy for opportunity) serve as controls for confounding factors.
In practice, real datasets fall short of this ideal in two ways. First, injury data is inconsistently recorded across eras, so missed games due to injury would serve as an imperfect proxy at best. Second, international players who were drafted but never played in the NBA often have incomplete career outcome data, which could introduce selection bias in later rounds.
Data would be collected from basketball-reference and official league records, stored as a tidy CSV where each player has a unique ID with draft and outcome fields merged into a single row. This structure enables statistical analyses such as correlation and regression to evaluate the relationship between draft position and career outcomes.

**Potential real datasets**

1. **Kaggle "20 Years of NBA Draft Data" by benwieland**  
Dataset link [https://www.kaggle.com/datasets/benwieland/nba-draft-data](https://www.kaggle.com/datasets/benwieland/nba-draft-data). This is the primary ready to use source because draft pick number, draft year, Win Shares, BPM, VORP, minutes, and games are already combined in one CSV. That structure makes it efficient for modeling how draft position relates to long term outcomes across the 2000 to 2018 draft window. The main limitation is that PER and All Star selections are both missing even though they are two of our three dependent variables.

2. **NBA API via the nba_api Python package**  
Project link [https://github.com/swar/nba_api](https://github.com/swar/nba_api). This source is used as a targeted supplement to obtain All Star appearance data that is missing from the Kaggle file. It can also provide player level identifiers and award records that help with accurate merging. We will not treat this API as the standalone core dataset for the project.

3. **Basketball Reference Draft Pages**  
Source link [https://www.basketball-reference.com/draft/](https://www.basketball-reference.com/draft/). We will use this source for verification and gap filling, especially for PER which is absent from the Kaggle dataset. It is also useful for checking draft level fields and resolving inconsistencies during merges. This role fits well because the Kaggle data was originally scraped from Basketball Reference.

We will use Basketball-Reference draft tables as the primary source, supplement with the Kaggle dataset for bulk stats, and use the NBA API to obtain or verify All-Star selections. When merging across sources, we will standardize player identifiers using Basketball-Reference player IDs or consistent name formatting and resolve any duplicates that arise.
