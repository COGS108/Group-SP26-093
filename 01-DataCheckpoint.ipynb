## Research Question

Is there a statistically significant relationship — measured through linear and logistic regression — between NBA draft position and player career success for players drafted from 2000 to 2018? We will measure career success using three outcomes: cumulative career Win Shares, career average Player Efficiency Rating (PER), and whether a player made at least one NBA All-Star team. The main predictor is draft position (pick number), and we will test whether higher picks are associated with stronger outcomes across these success metrics while controlling for minutes played and age at draft. This is a statistical inference question because we are evaluating the strength and significance of associations rather than building a pure prediction model.


## Background and Prior Work

The NBA Draft is one of the most important ways in which franchises add incoming talent, and draft pick order reflects each organization's perspective on a selected player's projected ability. Higher drafted players have greater expectations of development and thus greater potential for future success because of the extensive scouting resources and tactical planning that franchises invest in selecting the best possible prospect. Due to this, draft pick order is often used as an indicator of a player's perceived worth, so it is valuable to determine whether draft pick order can predict impactful success.

Basketball analytics provide a variety of statistics and measures that allow us to quantify player success based on contributions. Win Shares provide a way to estimate the number of games a franchise has won due to a specific player's effort. Per-minute productivity is summarized in PER (Player Efficiency Rating). Finally, All-Star selections represent a form of elite recognition of players, though they can also be influenced by outside factors including popularity and team success. Using multiple forms of measurement provides a more complete picture of how well a player performs during their career than using one metric alone.

Prior research about draft systems and future success has shown mixed findings across variables. According to a systematic review of North American professional sports, athletes selected in the first round typically perform better in both career length and statistical outcomes.1 In contrast, athletes selected in later rounds are less consistent statistically and therefore harder to evaluate. The study also reports inconsistencies across drafts, which suggests draft position is not a perfect indicator of long-term success.

A similar finding appears in research analyzing the relative value of NBA draft picks. The author concluded that although players selected earlier in the draft perform better on average than those selected later, the relationship is non-linear, meaning being a top pick does not guarantee elite professional performance.2 There is substantial variation in how strongly an individual athlete's performance correlates with draft position.

In addition to draft position and later performance, researchers have studied how general managers make draft decisions. Results indicate that while general managers may value attributes like height or speed more highly when drafting players, these perceptions do not always align with a player's realized contribution over time, such as long-term value or contract outcomes.3 These findings suggest inefficiencies in talent evaluation and show that factors such as player development opportunities, team environment, and access to playing time also shape career outcomes.

While these studies provide a foundation, each has meaningful limitations that motivate our work. Johnston et al. (2021) conduct a broad cross-sport review, which limits their ability to isolate NBA-specific dynamics such as the lottery system, rookie scale contracts, and the uneven distribution of development resources. Watave (2016) focuses on relative pick value but does not control for opportunity factors like minutes played, which can confound the relationship between draft position and performance. Glass (2023) examines general manager decision-making rather than long-term player outcomes directly. Critically, none of these studies simultaneously evaluate multiple career outcome metrics — Win Shares, PER, and All-Star appearances — within the same regression framework. Our study addresses these gaps by analyzing players drafted from 2000 to 2018 using a multi-outcome regression approach that controls for playing opportunity and age at draft, offering a more complete and methodologically unified picture of how draft position relates to NBA career success.

## Hypothesis


We hypothesize that earlier draft picks (lower pick numbers) will, on average, have higher career Win Shares, higher career average PER, and a greater probability of making at least one All-Star team than later picks. We also expect this relationship to be statistically significant but moderate in strength, since player development, injuries, team fit, and opportunity can substantially affect long-term outcomes beyond draft position alone. This expectation is consistent with how teams invest more scouting resources and organizational confidence in higher picks, while still acknowledging substantial variability in NBA career trajectories.

### Dataset 1 — NBA Draft History (2000–2018)

**Source**: [Basketball Reference Draft Pages](https://www.basketball-reference.com/draft/)

This dataset contains every player selected in the NBA Draft from 2000 through 2018 — a 19-year window that allows sufficient career time for even the latest cohort (drafted 2018) to have accumulated meaningful statistics by 2023–24. Each row represents one drafted player. The key independent variable is `pick_number` (overall draft position, 1–60), and the career outcome columns are `career_ws` (Win Shares), `career_bpm` (Box Plus/Minus), and `career_vorp` (Value Over Replacement Player).

**Win Shares** estimates the number of team wins attributable to a player over their career. It is context-dependent (a player on a winning team can accumulate more WS), but it scales with both quality and longevity, making it a useful cumulative success measure. A career total above 50 is typically associated with All-Star–caliber players; below 10 reflects a marginal NBA career.

**BPM** (Box Plus/Minus) is a per-100-possession impact estimate relative to league average. It is less sensitive to playing-time opportunity than Win Shares, making it a useful complementary measure of per-minute value. A BPM of 0 is exactly average; values above +3 indicate a meaningful positive impact.

**Important data limitations**:
- Players drafted but who never appeared in an NBA game have `career_games = 0` and all statistical columns missing or zero. These players are retained in the dataset with a flag (`never_played = True`) to avoid selection bias but are excluded from regression models that require career statistics.
- The "Round 1 / Round 2" label in Basketball Reference tables spans both rounds; the pick number is the global pick number (1–60 per year), which is the variable we use.
- PER is **not** available from these draft pages and is collected separately in Dataset 3.

The code below:
1. Scrapes all 19 draft-year pages (uses cached CSV if already downloaded)
2. Flattens multi-level column headers
3. Removes embedded header rows
4. Renames columns to snake_case
5. Converts numeric columns, flags players who never played
6. Saves raw data to `data/00-raw/` and cleaned data to `data/01-interim/`


```python
# ── Dataset 1: NBA Draft Data (2000–2018) ──────────────────────────────────────

RAW_DRAFT   = RAW_DIR       + 'nba_draft_2000_2018_raw.csv'
CLEAN_DRAFT = INTERIM_DIR   + 'nba_draft_2000_2018_clean.csv'

# ── 1a. Download (skip if cached) ─────────────────────────────────────────────

def flatten_draft_cols(df):
    """Flatten multi-level BBRef draft-page columns to single-level snake_case."""
    new_cols = []
    for top, bot in df.columns:
        if 'Unnamed' in str(top) or str(top) == '':
            new_cols.append(str(bot).strip())
        elif str(bot).strip() in ('Player', 'College'):
            new_cols.append(str(bot).strip())
        else:
            new_cols.append(f"{str(top).strip()}_{str(bot).strip()}")
    df.columns = new_cols
    return df

def scrape_draft_year(year):
    url = f'https://www.basketball-reference.com/draft/NBA_{year}.html'
    r = bball_ref_get(url)
    if r is None:
        return None
    tables = pd.read_html(StringIO(r.text))
    df = flatten_draft_cols(tables[0].copy())
    df['draft_year'] = year
    return df

if os.path.exists(RAW_DRAFT):
    print(f"Using cached raw draft data: {RAW_DRAFT}")
    draft_raw = pd.read_csv(RAW_DRAFT)
else:
    print("Scraping Basketball Reference draft pages (2000–2018)…")
    print("(~19 requests × 3 s delay ≈ 60 s)")
    frames = []
    for yr in range(2000, 2019):
        print(f"  {yr}", end=' ', flush=True)
        df = scrape_draft_year(yr)
        if df is not None:
            frames.append(df)
            print("✓")
        else:
            print("✗ (skipped)")
    draft_raw = pd.concat(frames, ignore_index=True)
    draft_raw.to_csv(RAW_DRAFT, index=False)
    print(f"\nSaved {len(draft_raw)} rows → {RAW_DRAFT}")

print(f"\nRaw shape: {draft_raw.shape}")
draft_raw.head(3)
```

### Dataset 2 — NBA All-Star Game Rosters (2000–2024)

**Source**: [Basketball Reference All-Star Game Pages](https://www.basketball-reference.com/allstar/)

The All-Star selection serves as a binary indicator of elite recognition: did a player ever receive an All-Star roster spot at any point in their career? We collect All-Star rosters for the years 2000–2024 to cover the full possible career window for players drafted from 2000 to 2018 (a 2018 draftee who made their first All-Star appearance in 2023 would be missed if we stopped at 2018).

**Why a binary flag**: Because All-Star selection is a categorical recognition (you either made it or you didn't), we model it as a binary outcome (`career_allstar = 1/0`) in a logistic regression. We also record the count of total selections for descriptive statistics.

**Limitations**:
- All-Star selections are influenced by fan voting, team success, and media market size in addition to pure performance, introducing noise.
- Injury replacements are included as valid selections (the player was awarded a spot).
- Name formatting inconsistencies between Basketball Reference datasets (e.g., accents, suffixes) require fuzzy matching for a small number of players.

The code below:
1. Scrapes All-Star game rosters for each year 2000–2024 (cached after first run)
2. Extracts player names from both conference (East/West) roster tables
3. Computes `allstar_count` and binary `career_allstar` per player
4. Saves to `data/01-interim/`


```python
# ── Dataset 2: NBA All-Star Rosters (2000–2024) ───────────────────────────────

RAW_ALLSTAR   = RAW_DIR     + 'nba_allstar_2000_2024_raw.csv'
CLEAN_ALLSTAR = INTERIM_DIR + 'nba_allstar_2000_2024_clean.csv'

# ── 2a. Download (skip if cached) ─────────────────────────────────────────────

SKIP_NAMES = {'Reserves', 'Team Totals', 'Starters'}

def scrape_allstar_year(year):
    url = f'https://www.basketball-reference.com/allstar/NBA_{year}.html'
    r = bball_ref_get(url)
    if r is None:
        return []
    tables = pd.read_html(StringIO(r.text))
    players = []
    # Tables 1 and 2 are the two conference rosters (skip table 0 = scoreline)
    for t in tables[1:3]:
        col = t.iloc[:, 0]
        for name in col:
            if pd.notna(name) and str(name).strip() not in SKIP_NAMES:
                players.append({'player_name': str(name).strip(), 'allstar_year': year})
    return players

if os.path.exists(RAW_ALLSTAR):
    print(f"Using cached raw All-Star data: {RAW_ALLSTAR}")
    allstar_raw = pd.read_csv(RAW_ALLSTAR)
else:
    print("Scraping Basketball Reference All-Star pages (2000–2024)…")
    print("(~25 requests × 3 s delay ≈ 75 s)")
    rows = []
    for yr in range(2000, 2025):
        print(f"  {yr}", end=' ', flush=True)
        entries = scrape_allstar_year(yr)
        rows.extend(entries)
        print(f"✓ ({len(entries)} players)")
    allstar_raw = pd.DataFrame(rows)
    allstar_raw.to_csv(RAW_ALLSTAR, index=False)
    print(f"\nSaved {len(allstar_raw)} rows → {RAW_ALLSTAR}")

print(f"Raw shape: {allstar_raw.shape}")
allstar_raw.head(6)
```

### Dataset 3 — Career Average Player Efficiency Rating (PER)

**Source**: [Basketball Reference Season Advanced Stats](https://www.basketball-reference.com/leagues/)

Player Efficiency Rating (PER) is a per-minute productivity measure developed by John Hollinger. It is normalized so that the league average is exactly 15.0 each season. Values above 20 correspond to All-Star–caliber performance; values below 10 indicate a below-average player. Unlike Win Shares, PER is scale-invariant with respect to minutes, making it a useful complement for comparing players who had very different roles and playing-time opportunities.

PER is not available on the draft-history pages and must be collected from per-season advanced stats tables. We scrape one page per season (2001–2023) and aggregate to a **games-weighted career average PER** for each player.

**Technical notes**:
- Basketball Reference's advanced stats pages have a `TOT` (total) row for players traded mid-season. We keep only the `TOT` row for such players to avoid double-counting.
- Players with fewer than 100 career games are flagged (`low_minutes = True`) because their PER estimates are less stable.
- We restrict to seasons 2001–2023 to cover the full career window for 2000–2018 draftees.

The code below:
1. Scrapes 23 season-advanced-stats pages (cached after first run)
2. Deduplicates traded-player rows, keeping only `TOT`
3. Aggregates to games-weighted career average PER per player
4. Saves to `data/01-interim/`


```python
# ── Dataset 3: Career Average PER from Season Advanced Stats ──────────────────

RAW_PER   = RAW_DIR     + 'nba_season_advanced_2001_2023_raw.csv'
CLEAN_PER = INTERIM_DIR + 'nba_career_avg_per_clean.csv'

# ── 3a. Download season-by-season advanced stats (skip if cached) ─────────────

def scrape_season_advanced(year):
    """Scrape NBA advanced stats for one season (year = ending calendar year)."""
    url = f'https://www.basketball-reference.com/leagues/NBA_{year}_advanced.html'
    r = bball_ref_get(url)
    if r is None:
        return None
    tables = pd.read_html(StringIO(r.text))
    df = tables[0].copy()
    # Remove repeated header rows
    df = df[df['Player'] != 'Player'].copy()
    df = df[df['Player'].notna()].copy()
    # For players traded mid-season, keep only TOT (aggregate) row
    # Sort so TOT comes first, then drop_duplicates keeps it
    team_col = 'Team' if 'Team' in df.columns else 'Tm'
    df = df.sort_values(team_col, key=lambda s: (s != 'TOT').astype(int))
    df = df.drop_duplicates(subset='Player', keep='first').copy()
    # Keep name, PER, games
    df = df[['Player', 'PER', 'G']].copy()
    df.columns = ['player_name', 'per', 'games']
    df['season'] = year
    return df

if os.path.exists(RAW_PER):
    print(f"Using cached raw PER data: {RAW_PER}")
    per_raw = pd.read_csv(RAW_PER)
else:
    print("Scraping Basketball Reference season advanced stats (2001–2023)…")
    print("(~23 requests × 3 s delay ≈ 70 s)")
    frames = []
    for yr in range(2001, 2024):
        print(f"  {yr}", end=' ', flush=True)
        df = scrape_season_advanced(yr)
        if df is not None:
            frames.append(df)
            print(f"✓ ({len(df)} players)")
        else:
            print("✗ (skipped)")
    per_raw = pd.concat(frames, ignore_index=True)
    per_raw.to_csv(RAW_PER, index=False)
    print(f"\nSaved {len(per_raw)} rows → {RAW_PER}")

print(f"Raw shape: {per_raw.shape}")
per_raw.head(3)
```

```python
# ── 3b. Aggregate to career average PER per player ────────────────────────────

per = per_raw.copy()
per['per']   = pd.to_numeric(per['per'],   errors='coerce')
per['games'] = pd.to_numeric(per['games'], errors='coerce')
per = per.dropna(subset=['per', 'games']).copy()

# Games-weighted career average PER
per['per_x_games'] = per['per'] * per['games']
per_agg = (
    per.groupby('player_name', as_index=False)
    .agg(career_total_games_per=('games', 'sum'),
         per_x_games_sum=('per_x_games', 'sum'))
)
per_agg['career_avg_per'] = per_agg['per_x_games_sum'] / per_agg['career_total_games_per']
per_agg = per_agg[['player_name', 'career_avg_per', 'career_total_games_per']].copy()
per_agg['low_minutes_flag'] = per_agg['career_total_games_per'] < 100

per_agg.to_csv(CLEAN_PER, index=False)
print(f"Unique players with career PER: {len(per_agg)}")
print(f"Players with <100 career games (unreliable PER): {per_agg['low_minutes_flag'].sum()}")
print(f"\nCareer avg PER distribution:")
print(per_agg['career_avg_per'].describe().round(2))
```

### Combined Dataset — Merging All Three Sources

We now merge all three datasets on `player_name` to produce the single analysis-ready file. The merge strategy is:

1. Start with **Dataset 1** (draft history, 2000–2018) as the left table — this defines the population of interest.
2. Left-join **Dataset 2** (All-Star flag) on `player_name`. Players with no All-Star appearance get `career_allstar = 0`.
3. Left-join **Dataset 3** (career avg PER) on `player_name`.

**Name matching note**: Basketball Reference uses consistent naming within the site, so most merges resolve cleanly. A small number of players with non-ASCII characters (accents, diacritics) were corrected via encoding normalization applied to the raw files.

**Duplicate player names**: Three pairs of distinct players share the same name in our dataset — Marcus Williams (drafted 2006 and 2007), Marcus Thornton (2009 and 2015), and Justin Jackson (2017 and 2018). Because Basketball Reference's per-season advanced stats pages do not distinguish between same-named players, PER data is assigned to the higher-games player in each pair (the one whose stats the seasonal data actually reflects). The second player in each pair (all three either never played or played very few games) has PER left as missing. None of these six players made an All-Star team, so the All-Star merge is unaffected.

The final file is written to `data/02-processed/nba_draft_career_stats.csv`.

```python
# ── Merge all three datasets ──────────────────────────────────────────────────

PROCESSED_FINAL = PROCESSED_DIR + 'nba_draft_career_stats.csv'

# Load cleaned interim files
draft   = pd.read_csv(CLEAN_DRAFT)
allstar = pd.read_csv(CLEAN_ALLSTAR)
per     = pd.read_csv(CLEAN_PER)

# Merge 1: draft + all-star
# None of the three duplicate-named pairs made the All-Star game, so this merge is clean.
merged = draft.merge(
    allstar[['player_name', 'career_allstar', 'allstar_count', 'first_allstar_year']],
    on='player_name', how='left'
)
merged['career_allstar'] = merged['career_allstar'].fillna(0).astype(int)
merged['allstar_count']  = merged['allstar_count'].fillna(0).astype(int)

# Merge 2: + PER (with duplicate-name handling)
# Three pairs of drafted players share a name: Marcus Williams (2006/2007),
# Marcus Thornton (2009/2015), Justin Jackson (2017/2018).
# BBRef season stats cannot separate same-named players, so we assign PER
# to the player with the most career games and leave the other's PER as NaN.
merged_sorted = merged.sort_values(
    ['player_name', 'career_games'],
    ascending=[True, False],
    na_position='last'
).copy()

merged_sorted = merged_sorted.merge(per, on='player_name', how='left')

per_cols = ['career_avg_per', 'career_total_games_per', 'low_minutes_flag']
dup_priority = merged_sorted.groupby('player_name').cumcount()
for col in per_cols:
    merged_sorted.loc[dup_priority > 0, col] = None

merged = merged_sorted.sort_values(['draft_year', 'pick_number']).reset_index(drop=True)

# Save final file
merged.to_csv(PROCESSED_FINAL, index=False)

print("=== Final combined dataset ===")
print(f"Shape: {merged.shape}")
print(f"Columns: {merged.columns.tolist()}")
print(f"\nMissing values per column:")
print(merged.isna().sum())
print(f"\nAll-Star players in draft dataset: {merged['career_allstar'].sum()}")
print(f"Players with PER data: {merged['career_avg_per'].notna().sum()}")
print(f"Players who never played in NBA: {merged['never_played'].sum()}")
print(f"\nDuplicate-name rows (PER assigned to higher-games player):")
dup_names = ['Marcus Williams', 'Marcus Thornton', 'Justin Jackson']
print(merged[merged['player_name'].isin(dup_names)][
    ['player_name', 'draft_year', 'pick_number', 'career_games', 'career_avg_per']
].to_string(index=False))
```
